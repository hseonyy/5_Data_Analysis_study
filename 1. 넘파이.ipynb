# **3. í–‰ë ¬ ì—°ì‚°**
* ë„˜íŒŒì´ì—ì„œëŠ” ë‹¤ì°¨ì› ë°°ì—´ì¸ ndarrayë¥¼ ì‚¬ìš©í•˜ì—¬ í–‰ë ¬ ì—°ì‚°ì„ ìˆ˜í–‰
* í–‰ë ¬ ì—°ì‚°ì€ ì„ í˜• ëŒ€ìˆ˜ì™€ ê´€ë ¨ì´ ê¹Šì–´ ë°ì´í„° ê³¼í•™, ë¨¸ì‹ ëŸ¬ë‹, í†µê²Œ ë“± ë‹¤ì–‘í•œ ë¶„ì•¼ì—ì„œ ì‚¬ìš©ë¨

```python
ndarr1 = np.array([[1, 2, 3],
                   [2, 3, 4]])
ndarr2 = np.array([[3, 4, 5],
                   [1, 2, 3]])
print(ndarr1.shape)
print(ndarr2.shape)
```

```python
# í–‰ë ¬ ë§ì…ˆ
print(ndarr1 + ndarr2)
```

```python
# í–‰ë ¬ ëº„ì…ˆ
print(ndarr1 - ndarr2)
```

```python
# í–‰ë ¬ ì›ì†Œë³„ ê³±ì…ˆ
print(ndarr1 * ndarr2)
```

```python
# í–‰ë ¬ ê³± (Dot Product)
print(ndarr1 @ ndarr2)  # ì—ëŸ¬
# ValueError: matmul: Input operand 1 has a mismatch in its core dimension 0, with gufunc signature (n?,k),(k,m?)->(n?,m?) (size 2 is different from 3)
# í–‰ë ¬ ê³±ì˜ ì¡°ê±´: ë§ë‹¿ëŠ” shapeê°€ ê°™ì•„ì•¼ í•¨, ë–¨ì–´ì ¸ ìˆëŠ” shapeê°€ ê²°ê³¼ í–‰ë ¬ì´ ë¨
# ì˜ˆ) (2, 3) @ (2, 3)  X
# ì˜ˆ) (3, 3) @ (3, 2)  O
```

```python
ndarr3 = np.array([[1, 2, 3],
                   [1, 2, 3],
                   [2, 3, 4]])
ndarr4 = np.array([[1, 2],
                   [3, 4],
                   [5, 6]])
print(ndarr3.shape)
print(ndarr4.shape)

print(ndarr3 @ ndarr4)
```

```python
# ë„˜íŒŒì´ë¡œ
print(np.dot(ndarr3, ndarr4))
```

```python
print((1*1 + 2*3 + 3*5), (1*2 + 2*4 + 3*6))
print((1*1 + 2*3 + 3*5), (1*2 + 2*4 + 3*6))
print((2*1 + 3*3 + 4*5), (2*2 + 3*4 + 4*6))
```

```python
# ì „ì¹˜ í–‰ë ¬
# ê¸°ì¡´ í–‰ë ¬ì˜ í–‰ê³¼ ì—´ì„ ë°”ê¾¼ ìƒˆë¡œìš´ í–‰ë ¬
# T
ndarr1 = np.array([[1, 2, 3],
                   [2, 3, 4]])
print(ndarr1)
print(ndarr1.T)
```

```python
# ì—­í–‰ë ¬
# ì£¼ì–´ì§„ ì •ì‚¬ê° í–‰ë ¬ì— ëŒ€í•œ ê³±ì…ˆ ì—°ì‚°ìœ¼ë¡œ ë‹¨ìœ„ í–‰ë ¬ì„ ì–»ì„ ìˆ˜ ìˆëŠ” í–‰ë ¬
# ë‹¨ìœ„ í–‰ë ¬: ì£¼ëŒ€ê°ì„ ì˜ ì›ì†Œê°€ ëª¨ë‘ 1ì´ê³ , ë‚˜ë¨¸ì§€ ì›ì†Œê°€ ëª¨ë‘ 0ì¸ ì •ì‚¬ê°í˜• í–‰ë ¬
arr = np.array([[1, 2], [3, 4]])
print(np.linalg.inv(arr))

print(arr @ np.linalg.inv(arr))
```
